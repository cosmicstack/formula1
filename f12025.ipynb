```python
current_points = {
    "NOR": 408,
    "VER": 396,
    "PIA": 392,
    "RUS": 309,
    "LEC": 230,
    "HAM": 152,
    "ANT": 150,
    "ALB": 73,
    "SAI": 64,
    "HAD": 51,
    # "HUL": 49,
    # "ALO": 48,
    # "BEA": 41,
    # "LAW": 38,
    # "TSU": 33,
    # "OCO": 32,
    # "STR": 32,
    # "GAS": 22,
    # "BOR": 19,
    # "COL": 0,
    # "DOO": 0
}
```

```python
winning_points = [25, 18, 15, 12, 10, 8, 6, 4, 2, 1]
```

```python
len(winning_points)
```

```python
possible_points = np.array(list(permutations(winning_points)))
```

```python
possible_points.shape
```

```python
final_standings = possible_points + np.array(list(current_points.values()))
```

```python
df_argsort = 10 - pd.DataFrame(final_standings.argsort())
```

```python
predicted = df_argsort.groupby(df_argsort.columns.tolist(), as_index=False).size().rename(columns=dict(zip(np.arange(0,10), current_points.keys())))
```

```python
predicted['prob'] = predicted['size']/predicted['size'].sum()
```
